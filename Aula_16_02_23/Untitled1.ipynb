```python
import numpy as np
import cv2 as cv
import matplotlib as mpl
from matplotlib import pyplot as plt
import math
```

```python
img_BGR = cv.imread("Folha.jpg")
img_RGB = cv.cvtColor(img_BGR,cv.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
#faz split das bandas da imagem RGB
R, G, B = cv.split(img_RGB)
plt.imshow(R, cmap='gray')
plt.show()
plt.imshow(G, cmap='Wistia')
plt.show()
plt.imshow(B, cmap='Pastel2')
plt.show()
```

```python
#obtem as dimenses da imagem (nmero de linhas e nmero de colunas)
num_lin = img_RGB.shape[0]
num_col = img_RGB.shape[1]
#cria uma matriz preenchida com zeros - nela sero armazenados os nveis de cinzas das mdias computadas (R+G+B)/3
img_gray=np.zeros((num_lin,num_col), dtype="uint8")
for l in range (num_lin):
    for c in range (num_col):
        (r, g, b) = img_RGB[l,c]
        img_gray[l,c] = ((int(r))+int(g)+int(b))//3
plt.imshow(img_gray,cmap='gray')
```

```python
import cv2
imagem = cv2.imread("cogumelo.png")
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

for y in range(0, image.shape[0]):
    for x in range(0, image.shape[1]):
        if image[y, x] == 255:
            image[y, x] == 0
    else: 
        image[y,x] = 255
        
plt.imshow(image, interpolation="none", cmap="gray")
plt.show()
```
